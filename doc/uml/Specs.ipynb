# Spécifications Fonctionnelles – Système de Gestion du Prix Goncourt

### 1. Objectifs
- Permettre la gestion des livres candidats au Prix Goncourt.
- Suivre les sélections successives (1 ère, 2 ème, finale).
- Gérer les membres du jury et leurs votes.
- Produire des résultats fiables (livre gagnant, statistiques des votes).

### 2. Périmètre fonctionnel
#### 2.1 Gestion des livres
- Attributs : titre, résumé, date de parution, nbPages, ISBN, prix éditeur.
- Chaque livre est relié à un auteur et un éditeur.
- Chaque livre comporte zéro, un ou plusieurs personnages.

#### 2.2 Gestion des auteurs
- Attributs : nom, prénom, biographie (en option).
- Un auteur peut avoir écrit plusieurs livres.

#### 2.3 Gestion des éditeurs
- Attributs : nom, adresse.
- Un éditeur peut publier plusieurs livres.

#### 2.4 Gestion des personnages
- Attributs : nom, rôle.
- Chaque personnage est lié à un seul livre dans notre application.

#### 2.5 Gestion des sélections
- Associer plusieurs livres à une sélection.
- Un livre peut apparaître dans plusieurs sélections.

#### 2.6 Gestion des membres du jury
- Attributs : nom, rôle, ainsi que l'identifiant et mot de passe optionnels pour le moment.
- Un membre peut voter pour un seul livre par selection.

#### 2.7 Gestion des votes
- Chaque vote est lié à un membre du jury et à un livre.
- Un livre peut recevoir plusieurs votes.
- Calculer automatiquement le total des voix par livre.

#### 3. Règles de gestion
- Un livre doit avoir exactement un auteur et exactement un éditeur.
- Une sélection doit contenir au moins un livre.
- Un vote est toujours relié à un seul livre et un seul membre du jury.
- Le gagnant est le livre ayant obtenu le plus de voix dans la sélection finale.

#### 4. Rapports et sorties
- Liste des livres par sélection.
- Liste des votes.
- Statistiques : total des voix par livre, par sélection.
- Détermination automatique du gagnant.
